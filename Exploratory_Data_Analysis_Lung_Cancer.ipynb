**Measure of Shape**

**Skewness**

1. Negative Skewness => Mean < Median < Mode
2. Zero Skewness => Mean = Median = Mode
3. Positive Skewness => Mean > Median > Mode

Aturan praktis untuk nilai skewness adalah:

Jika skewness berada di antara -0,5 dan 0,5, data cukup simetris.

Jika skewness berada di antara -1 dan - 0,5 atau di antara 0,5 dan 1, data cukup miring.

Jika skewness kurang dari -1 atau lebih besar dari 1, data sangat miring.

```python
print(f'Skewness of LUNG_CANCER: {df['LUNG_CANCER'].skew()}')
```

**Penjelasan:**

Skewness pada variabel target kita ternyata kurang dari -1. Jadi, kita dapat simpulkan bahwa variabel target kita memiliki kemiringan yang sangat miring yakni skewness negative atau skew kiri.

**Kurtosis**

Kurva mesokurtik

Distribusi normal memiliki kurtosis tepat 3 (kurtosis berlebih tepat 0). Setiap distribusi dengan kurtosis ≈3 (kelebihan kurtosis ≈ 0) disebut mesokurtik.

Kurva platykurtik

Distribusi dengan kurtosis < 3 (kelebihan kurtosis < 0) disebut platykurtik. Dibandingkan dengan distribusi normal, puncak pusatnya lebih rendah dan lebih luas, dan ekornya lebih pendek dan lebih tipis.

Kurva leptokurtik

Distribusi dengan kurtosis > 3 (kelebihan kurtosis > 0) disebut leptokurtik. Dibandingkan dengan distribusi normal, puncak pusatnya lebih tinggi dan lebih tajam, dan ekornya lebih panjang dan lebih gemuk.


```python
print(f'Kurtosis-LUNG CANCER: {df['LUNG_CANCER'].kurt()}')
```

**Penjelasan:**

Kurtosis pada kolom LUNG_CANCER lebih besar dari pada 3, yakni termasuk kedalam Leprokurtik. Data ini memiliki puncak yang lebih tajam dan ekor yang lebih tebal dibandingkan dengan distribusi normal, yang mungkin menunjukkan adanya nilai-nilai ekstrem atau outliers.

**Plotting**

```python
plt.figure(figsize=(8,6))
sns.distplot(df['LUNG_CANCER'], kde=False, fit=st.norm)
plt.title('Normal fit')
plt.show()
```

**Penjelasan:**

Ini adalah grafik LUNG_CANCER dengan menggunakan variabel normal distribusi

```python
plt.figure(figsize=(8,6))
sns.distplot(df['LUNG_CANCER'], kde=False, fit=st.lognorm)
plt.title('Log Normal fit')
plt.show()
```

**Penjelasan:**

Ini adalah grafik LUNG_CANCER dengan lognormal distribusi

**Kesimpulan kedua grafik:**
Distribusi ini bersifat leptokurtik, dengan mayoritas data terkonsentrasi di sekitar nilai 1. Hal ini menunjukkan bahwa data memiliki pola dominan di satu kategori (atau kelompok) dengan sedikit nilai lain yang berbeda. Jika tujuan analisis adalah memahami variabilitas atau distribusi, data ini mungkin memerlukan perhatian khusus seperti pengelompokan ulang atau transformasi untuk analisis yang lebih mendalam.

**Penjelasan:**

Boxplot ini menampilkan distribusi usia pasien berdasarkan status kanker paru-paru. Terdapat dua kelompok yang ditampilkan:

Kelompok 1: Pasien tanpa kanker paru-paru.

Kelompok 2: Pasien dengan kanker paru-paru.

Elemen Utama Boxplot:

Median (Garis di dalam kotak):Menunjukkan nilai tengah dari distribusi usia pada masing-masing kelompok.

1. Kelompok tanpa kanker paru-paru (1): Median sekitar 60 tahun.
2. Kelompok dengan kanker paru-paru (2): Median sedikit lebih tinggi, sekitar 65 tahun.

IQR (Interquartile Range - Kotak):Menunjukkan rentang antara kuartil pertama (Q1 - 25%) dan kuartil ketiga (Q3 - 75%).

1. Kelompok tanpa kanker paru-paru: Rentang usia sekitar 50-70 tahun.
2. Kelompok dengan kanker paru-paru: Rentang usia sekitar 55-75 tahun.

Whiskers (Garis horizontal di luar kotak): Menunjukkan jangkauan data di luar IQR, biasanya sampai 1.5 kali IQR. Data di luar whiskers dianggap sebagai outlier.

1. Outlier terlihat di kedua kelompok (terutama pada usia yang lebih muda dan lebih tua).

Outliers (Lingkaran di luar whiskers): Data yang sangat jauh dari IQR dan mungkin memerlukan perhatian lebih lanjut.

1. Pada kelompok tanpa kanker paru-paru, terdapat outlier di usia sekitar 20 tahun.
2. Pada kelompok dengan kanker paru-paru, terdapat outlier di sekitar usia 85 tahun.
